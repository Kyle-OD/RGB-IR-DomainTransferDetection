```python
import matplotlib.pyplot as plt
```

```python
epochs = []
iters = []
times = []
data_losses = []
G_GAN_losses = []
G_L1_losses = []
D_real_losses = []
D_fake_losses = []
```

```python
with open('pytorch-CycleGAN-and-pix2pix/checkpoints/LLVIP2/loss_log.txt', 'r') as file:
    lines = file.readlines()
```

```python
parts = lines[1].split()
```

```python
parts
```

```python
for line in lines[1:]:
    parts = line.split()
    epochs.append(int(parts[1].split(',')[0]))
    iters.append(int(parts[3].split(',')[0]))
    times.append(float(parts[5].split(',')[0]))
    data_losses.append(float(parts[7].split(')')[0]))
    G_GAN_losses.append(float(parts[9]))
    G_L1_losses.append(float(parts[11]))
    D_real_losses.append(float(parts[13]))
    D_fake_losses.append(float(parts[15]))
```

```python
plt.figure(figsize=(10, 6))

plt.plot(G_GAN_losses, label='G_GAN Loss')
plt.plot(G_L1_losses, label='G_L1 Loss')
plt.plot(D_real_losses, label='D_real Loss')
plt.plot(D_fake_losses, label='D_fake Loss')

plt.xlabel('Iterations')
plt.ylabel('Loss')
plt.title('Losses Over Time')
plt.legend()
plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(10, 6))

plt.plot(data_losses, label='Total Loss')

plt.xlabel('Iterations')
plt.ylabel('Loss')
plt.title('Loss Over Time')
plt.legend()
plt.grid(True)

plt.show()
```
